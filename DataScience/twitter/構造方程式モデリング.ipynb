```python
def draw_graph(cm, threshold=0.3):
    edges = np.where(np.abs(cm) > threshold)
    edges = [[cm.index[i], cm.index[j]] for i, j in zip(edges[0], edges[1]) if i > j]

    g = Graph(format='png')
    for k in range(cm.shape[0]):
        g.node(cm.index[k])

    for i, j in edges:
        g.edge(j, i)

    g.render(filename='data/test', format='png', cleanup=True, directory=None)
    display(Image.open('data/test.png'))
```

```python
from sklearn.decomposition import FactorAnalysis
from sklearn.preprocessing import StandardScaler
```

```python
# Ê®ôÊ∫ñÂåñ
tw_tar = tw[tar_cols].copy()
sc = StandardScaler()
sc.fit(tw_tar)
z = sc.transform(tw_tar)
tw_sc = pd.DataFrame(data=z,columns=tar_cols)
```

```python
print(tw_sc.shape)
tw_sc.head()
```

```python
n_factors=3
fa = FactorAnalysis(n_components=n_factors,rotation='varimax', random_state=57)
fa.fit(tw_sc)

# Âõ†Â≠êË≤†Ëç∑Èáè
df_factor_loading = pd.DataFrame(
  fa.components_.T, columns=[
    'factor_{}'.format(i) for i in range(n_factors)
  ], index=tw_sc.columns
)

# Âõ†Â≠êÂæóÁÇπ
factor_scores = fa.transform(tw_sc)
df_factor_scores = pd.DataFrame(
  factor_scores, columns=['factor_{}'.format(i) for i in range(n_factors)], index=tw_sc.index
)
```

### Kmeans„ÇØ„É©„Çπ„Çø„É™„É≥„Ç∞ & ‰∏ªÊàêÂàÜÂàÜÊûê

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# kmeans
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit(tw_sc)
tw_sc["km_cluster"] = clusters.labels_
print(tw_sc["km_cluster"].unique())
tw_sc = tw_sc.reset_index(drop=True)

# pca
X = tw_sc
pca = PCA(n_components=2)
pca.fit(X)
x_pca = pca.transform(X)
pca_df = pd.DataFrame(x_pca)
pca_df["km_cluster"] = tw_sc["km_cluster"]
```

```python
for i in tw_sc["km_cluster"].unique():
    tmp = pca_df.loc[pca_df["km_cluster"]==i]
    plt.scatter(tmp[0], tmp[1], label=i)
plt.legend()
plt.xlabel('‰∏ªÊàêÂàÜ1')
plt.ylabel('‰∏ªÊàêÂàÜ2')
plt.title('kmeans + pca')
```

# ÊßãÈÄ†ÊñπÁ®ãÂºè„É¢„Éá„É™„É≥„Ç∞ÔºàSEM: Structural Equation ModelingÔºâ

- =~ Ê∏¨ÂÆöÊºîÁÆóÂ≠ê
- ~ ÂõûÂ∏∞ÊºîÁÆóÂ≠ê
- ~~ ÂàÜÊï£ÊºîÁÆóÂ≠ê

```python
mod = """\
engagement =~ „Ç§„É≥„Éó„É¨„ÉÉ„Ç∑„Éß„É≥ + „ÅÑ„ÅÑ„Å≠ + „É™„ÉÑ„Ç§„Éº„Éà
media =~ media_exist + „É°„Éá„Ç£„Ç¢„ÅÆÂÜçÁîüÊï∞ + „É°„Éá„Ç£„Ç¢„ÅÆ„Ç®„É≥„Ç≤„Éº„Ç∏„É°„É≥„ÉàÊï∞
reply =~ reply_flg + Ëøî‰ø°
click =~ „É¶„Éº„Ç∂„Éº„Éó„É≠„Éï„Ç£„Éº„É´„ÇØ„É™„ÉÉ„ÇØ + „Éè„ÉÉ„Ç∑„É•„Çø„Ç∞„ÇØ„É™„ÉÉ„ÇØ + Ë©≥Á¥∞„ÇØ„É™„ÉÉ„ÇØ
engagement ~ media
engagement ~ click
reply ~ engagement
engagement ~~ media
„Ç§„É≥„Éó„É¨„ÉÉ„Ç∑„Éß„É≥ ~~ „É°„Éá„Ç£„Ç¢„ÅÆÂÜçÁîüÊï∞
„ÅÑ„ÅÑ„Å≠ ~~ Ë©≥Á¥∞„ÇØ„É™„ÉÉ„ÇØ
"""
```

```python
model = semopy.Model(mod)
```

```python
tar_cols = ['„Ç§„É≥„Éó„É¨„ÉÉ„Ç∑„Éß„É≥', '„ÅÑ„ÅÑ„Å≠', '„É™„ÉÑ„Ç§„Éº„Éà', 
            'media_exist', '„É°„Éá„Ç£„Ç¢„ÅÆÂÜçÁîüÊï∞', '„É°„Éá„Ç£„Ç¢„ÅÆ„Ç®„É≥„Ç≤„Éº„Ç∏„É°„É≥„ÉàÊï∞', 
            'reply_flg', 'Ëøî‰ø°',
            '„É¶„Éº„Ç∂„Éº„Éó„É≠„Éï„Ç£„Éº„É´„ÇØ„É™„ÉÉ„ÇØ','„Éè„ÉÉ„Ç∑„É•„Çø„Ç∞„ÇØ„É™„ÉÉ„ÇØ', 'Ë©≥Á¥∞„ÇØ„É™„ÉÉ„ÇØ']

df_tw = tw[tar_cols].copy()
df_tw = df_tw.astype(np.int64)
```
