## ç„¡æ„å‘³ãªãƒ‡ãƒ¼ã‚¿å‰Šé™¤

```python
tw = tw.replace({'-': pd.np.nan,'NaN': pd.np.nan, 'nan': pd.np.nan})
tw = tw.dropna(how='all',axis=1)
print(tw.shape)
tw.head(1)
```

```python
# é‡è¤‡ã—ã¦ã„ã‚‹ãƒ„ã‚¤ãƒ¼ãƒˆã‚’ãƒ„ã‚¤ãƒ¼ãƒˆIDã«ã¤ã„ã¦ä¸€æ„ã«å¤‰æ›´
tw = tw.loc[~(tw.duplicated(subset=['ãƒ„ã‚¤ãƒ¼ãƒˆID']))]
tw = tw.reset_index(drop=True)
tw = tw.fillna(0)
tw.head(1)
```

```python
# å…¨ã¦ãŒ0ã®åˆ—ã‚’ç¢ºèª
all_0_cols = []

for col in tw.columns:
  if (tw[col] == 0).all() == True:
    all_0_cols.append(col)
    
print(all_0_cols)
```

```python
tw = tw.drop(columns=all_0_cols)
tw = tw.sort_values('æ™‚é–“', ascending=False)
tw = tw.reset_index(drop=True)

print(tw.shape)
tw.head(2)
```

## ãƒ‡ãƒ¼ã‚¿æ•´ç†

```python
tw.isnull().sum()
```

```python
# ã‚«ãƒ©ãƒ è¿½åŠ 

tw['media_exist'] = tw['ãƒ„ã‚¤ãƒ¼ãƒˆæœ¬æ–‡'].str.contains('t.co')
tw['reply_flg'] = tw['ãƒ„ã‚¤ãƒ¼ãƒˆæœ¬æ–‡'].str.contains('@')

import re

def format_text(text):
  text=re.sub(r'https?://[\w/:%#\$&\?\(\)~\.=\+\-â€¦]+', "", text)
  text=re.sub(r'[!-~]', "", text)#åŠè§’è¨˜å·,æ•°å­—,è‹±å­—
  text=re.sub(r'[ï¸°-ï¼ ]', "", text)#å…¨è§’è¨˜å·
  text=re.sub('\n', " ", text)#æ”¹è¡Œæ–‡å­—

  return text

tw['ãƒ„ã‚¤ãƒ¼ãƒˆæœ¬æ–‡_x'] = tw['ãƒ„ã‚¤ãƒ¼ãƒˆæœ¬æ–‡'].apply(lambda s: format_text(s))
tw['æ–‡å­—æ•°'] = tw['ãƒ„ã‚¤ãƒ¼ãƒˆæœ¬æ–‡_x'].str.len()

tw['æ™‚é–“'] = pd.to_datetime(tw['æ™‚é–“'])
tw['YMD'] = tw['æ™‚é–“'].dt.strftime('%Y%m%d')
tw['YEAR'] = tw['æ™‚é–“'].dt.year
tw['MONTH'] = tw['æ™‚é–“'].dt.month
tw['DAY'] = tw['æ™‚é–“'].dt.day
tw['TIME'] = tw['æ™‚é–“'].dt.hour
tw['WEEKDAY'] = tw['æ™‚é–“'].dt.weekday
tw['WEEK'] = tw['æ™‚é–“'].dt.week

tw = tw.drop(columns='ãƒ„ã‚¤ãƒ¼ãƒˆã®å›ºå®šãƒªãƒ³ã‚¯')
```
