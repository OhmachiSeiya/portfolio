### 相関分析

```python
tar_cols = ['いいね', 'リツイート', 'インプレッション','返信',
            'ユーザープロフィールクリック', 'URLクリック数', '詳細クリック',
            'フォローしている', 'メディアの再生数', '文字数']
# sns.pairplot(tw[tar_cols],
#              hue='フォローしている', kind='reg')
# plt.savefig('data/pairplot_byFollow.png')

fig = plt.figure(figsize=(24,24))
axes = fig.add_subplot(111)
axes.axis("off")
plt.grid(False) 

axes.imshow(mpimg.imread("data/pairplot_byFollow.png"))
```

```python
# 偏相関行列

r = tw[tar_cols].corr().values #相関行列
r_inv = np.linalg.inv(r) #相関行列の逆行列

rows = r_inv.shape[0]
regu_1 = 1 / np.sqrt(np.diag(r_inv))
regu_2 = np.repeat(regu_1, rows).reshape(rows, rows)
pcor = (-r_inv) * regu_1 * regu_2
np.fill_diagonal(pcor, 1)
```

```python
df_pcor = pd.DataFrame(pcor,columns=tar_cols, index=tar_cols)
df_pcor
```

```python
from graphviz import Graph
from PIL import Image

def draw_graph(cm, threshold=0.3):
    edges = np.where(np.abs(cm) > threshold)
    edges = [[cm.index[i], cm.index[j]] for i, j in zip(edges[0], edges[1]) if i > j]

    g = Graph(format='png')
    for k in range(cm.shape[0]):
        g.node(cm.index[k])

    for i, j in edges:
        g.edge(j, i)

    g.render(filename='data/test', format='png', cleanup=True, directory=None)
    display(Image.open('data/test.png'))
```

```python
draw_graph(tw[tar_cols].corr())
```

### Kmeansクラスタリング & 主成分分析

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# kmeans
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit(tw_sc)
tw_sc["km_cluster"] = clusters.labels_
print(tw_sc["km_cluster"].unique())
tw_sc = tw_sc.reset_index(drop=True)

# pca
X = tw_sc
pca = PCA(n_components=2)
pca.fit(X)
x_pca = pca.transform(X)
pca_df = pd.DataFrame(x_pca)
pca_df["km_cluster"] = tw_sc["km_cluster"]
```

```python
for i in tw_sc["km_cluster"].unique():
    tmp = pca_df.loc[pca_df["km_cluster"]==i]
    plt.scatter(tmp[0], tmp[1], label=i)
plt.legend()
plt.xlabel('主成分1')
plt.ylabel('主成分2')
plt.title('kmeans + pca')
```

```python
# 0を含む分布と含まない分布を比較
cols = ['リツイート', 'エンゲージメント', 'ユーザープロフィールクリック', '詳細クリック', 'メディアの再生数']
# 分布
for c in cols:
  tmp_data = tw.loc[tw[c] > 0, c]
  print("0の数: ", len(tmp_data.loc[tmp_data == 0]))
  
  fig = plt.figure(figsize=(14,4))
  fig.suptitle(c + " 0含む | 0含まない", fontsize=16)
  
  ax1 = fig.add_subplot(121)
  sns.histplot(tw[c], ax=ax1)
  ax1t = ax1.twinx()
  sns.kdeplot(tw[c], ax=ax1t)
  
  ax2 = fig.add_subplot(122)
  sns.histplot(tmp_data, ax=ax2)
  ax2t = ax2.twinx()
  sns.kdeplot(tmp_data, ax=ax2t)
  
  plt.tight_layout()
  plt.show()
```

```python
cols = ['リツイート', 'エンゲージメント', 'ユーザープロフィールクリック', '詳細クリック', 'メディアの再生数']
# 分布
for c in cols:
  tmp_data = tw[c].apply(np.log1p)
  
  fig = plt.figure(figsize=(14,4))
  fig.suptitle(c + " そのまま | 対数変換", fontsize=16)
  
  ax1 = fig.add_subplot(121)
  sns.histplot(tw[c], ax=ax1)
  ax1t = ax1.twinx()
  sns.kdeplot(tw[c], ax=ax1t)
  
  ax2 = fig.add_subplot(122)
  sns.histplot(tmp_data, ax=ax2)
  ax2t = ax2.twinx()
  sns.kdeplot(tmp_data, ax=ax2t)
  
  plt.tight_layout()
  plt.show()
```

```python
# ダミー変数化
df_w = pd.get_dummies(tw.WEEKDAY, dummy_na=True, prefix='WDAY',drop_first=True)
df_t = pd.get_dummies(tw.TIME, dummy_na=True, prefix='TIME',drop_first=True)
df_reg = pd.concat([tw, df_w, df_t], axis=1)
```

```python
df_reg.columns
```

```python
for c in df_reg.loc[:,df_reg.columns.str.contains('WDAY|TIME')].columns:
  df_reg = df_reg.rename(columns={c:c[:-2]})
```

```python
df_reg.loc[:,df_reg.columns.str.contains('WDAY|TIME')].columns
```

```python
df_reg.columns[-31:-25]
```

```python
df_reg.columns[-24:-1]
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# 重回帰分析 ステップワイズ法
# 動かせる説明変数
reg = "いいね ~ WDAY_1 + WDAY_2 + WDAY_3 + WDAY_4 + WDAY_5 + WDAY_6"
model = smf.ols(formula=reg, data=df_reg)
result = model.fit()
```

```python
result.summary()
```

```python
import statsmodels.discrete.discrete_model as sdm
# ロジスティック回帰
# Fit and summarize OLS model

reg = "フォローしている ~ 文字数 + reply_flg + media_exist + WDAY_1 + WDAY_2 + WDAY_3 + WDAY_4 + WDAY_5 + WDAY_6"
model = smf.glm(formula=reg, data=df_reg, family= sm.families.Binomial() )
res = model.fit() 
```
